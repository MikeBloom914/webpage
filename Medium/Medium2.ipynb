```python
stock = 'NVDA'
#### Write start and end dates in a string in the format YYYY-MM-DDDD ####
start_date = '2016-01-01'
end_date = '2018-01-01'
```

```python
deep_link = 'https://api.coindesk.com/v1/bpi/historical/close.json?start={start}&end={end}'.format(start=start_date, end=end_date)
response = json.loads(requests.get(deep_link).text)
```

```python
deep_link = 'https://www.quandl.com/api/v3/datasets/WIKI/{stock}.json?start_date={start_date}&end_date={end_date}'.format(stock=stock, start_date=start_date, end_date=end_date)

response_two = json.loads(requests.get(deep_link).text)
response_two = (response_two['dataset'])
```

```python
a = []
x = [[k, v] for k, v in response_two.items()]
y = x[17][1]
for i in y:
    key = (i[0])
    value = (i[4])
    a.append((key, value))

```

```python
stock = sorted(a)
bipdic = dict(stock)
```

```python
with open('bit_dates.csv', 'r+') as f:
    dates = csv.writer(f)
    dates.writerow(["Date", "Bitcoin"])
    for key, value in stock_dict.items():
        dates.writerow([key, value])


with open('stock.csv', 'r+') as g:
    dates = csv.writer(g)
    dates.writerow(["Date", "Stock"])
    for key, value in bipdic.items():
        dates.writerow([key, value])
```

```python
a = pd.read_csv("bit_dates.csv")
b = pd.read_csv("stock.csv")
b = b.dropna(axis=1)
df = a.merge(b, on='Date')
df.to_csv("combined.csv", index=False)
corr = df['Stock'].corr(df['Bitcoin'])
print(df.head())
print(round(corr, 4))
```
