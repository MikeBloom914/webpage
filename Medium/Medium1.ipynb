```python
#!usr/bin/env python3
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import quandl
dataframe = quandl.get("WIKI/NVDA", start_date="2016-01-01", end_date="2018-01-01")
```

```python
dataframe
```

```python
# Let's check to see what is the index column in our chart #
dataframe.index
```

```python
df = dataframe[['Open','High','Low','Close','Volume']]
df
```

```python
df['Change'] = df.Open - df.Close
```

```python
df.head()
```

```python
round(df['Change'].head().max(), 1)
```

```python
thirty_day_ma = df['Close'].rolling(window=30).mean()
one_hundred_day_ma = df['Close'].rolling(window=100).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(df.index, df['Close'], label='NVDA')
ax.plot(thirty_day_ma.index, thirty_day_ma, label='30 day rolling MA')
ax.plot(one_hundred_day_ma.index, one_hundred_day_ma, label='100 day rolling MA')

ax.set_xlabel('Date')
ax.set_ylabel('Closing price ($)')
ax.legend()

```
